```python
from azureml import Workspace
ws = Workspace()
experiment = ws.experiments['ae4c6f4dbab04496af447af010096b90.f-id.ee10dc5ca11a4aea9ede144284f8c3e5']
ds = experiment.get_intermediate_dataset(
    node_id='509577d4-c117-4cbf-ab26-f5a8b0a6810a-57475',
    port_name='Results dataset',
    data_type_id='GenericCSV'
)
frame = ds.to_dataframe()
```

```python
frame.describe()
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8,6)) # define plot area
ax = fig.gca() # define axis    
frame['ArrDelay'].plot.hist(ax = ax, bins = 30) # Use the plot.hist method on subset of the data frame
ax.set_title('Histogram of ArrDelay') # Give the plot a main title
ax.set_xlabel('ArrDelay') # Set text for the x axis
ax.set_ylabel('Number of flights')# Set text for y axis
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8,6)) # define plot area
ax = fig.gca() # define axis    
frame[['ArrDelay']].boxplot(ax = ax) # Use the plot.bar method on the new data frame
ax.set_title('Box plots ofArrDelay') # Give the plot a main title
ax.set_xlabel('ArrDelay') # Set text for the x axis
ax.set_ylabel('Number of flights')# Set text for y axis
```
